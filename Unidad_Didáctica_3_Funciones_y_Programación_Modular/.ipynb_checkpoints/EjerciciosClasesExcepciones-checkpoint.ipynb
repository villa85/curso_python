### Excepciones

Una excepción, es una señal que ha ocurrido una condición, que no se puede manejar fácilmente usando el flujo de control normal de un programa Python. Las excepciones a menudo se definen como "errores", pero no siempre es así. Todos los errores en Python se tratan mediante excepciones, pero no todas las excepciones son errores.

```python
try:
   <try clause code block>
except <ErrorType>:
   <exception handler code block>
```

```python
palabra = "Buenas noches"
palabra[13]
```

```python
try:
    palabra = "Buenas noches"
    palabra[13]
    print("Esto no se imprime")
except Exception:    # gestiona todos los errores de tiempo de ejecución
    print("Ha ocurrido un error")

print("El programa continuará ejecutándose...")
```

```python
items = ['a', 'b']
third = items[2]  # número de elementos de la lista
```

```python
try:
    items = ['a', 'b']
    third = items[2]
    print("Esto no se imprime")
except Exception:    # gestiona todos los errores de tiempo de ejecución
    print("Ha ocurrido un error")

print("El programa continuará ejecutándose...")
```

```python
try:
    items = ['a', 'b']
    third = items[2]
    print("Esto no se imprime")
except IndexError:    # gestiona sólo los IndexError, podemos definir una clase de errores más restringida
    print("Ha ocurrido un error, esta buscando una posición o elemento que no existe")

print("El programa continuará ejecutándose...")
```

```python
# OSERROR
import os
import sys

try: 
    # Si el fichero no existe se genera un IOError 
    archivo = 'GFG.txt'
    f = open(archivo, "r") 
    texto = f.read() 
    f.close() 
except IOError as error: 
    print('Problema con el fichero: {}.  {}'.format(archivo, error) )
except:  # si es un error diferente a IOError
    print("Error inesperado:", sys.exc_info()[0])
    raise
```

### Cuando utilizar Try/Except
La razón para usar try / except es cuando tiene un bloque de código para ejecutar que a veces se ejecutará correctamente y otras no, dependiendo de las condiciones que no puede prever en el momento en que está escribiendo el código.


```python
import os
PATH_FILE = os.getcwd() # el directorio actual
print(PATH_FILE)

try:
    PATH_IMAGES = PATH_FILE + "\images"
    os.chdir(PATH_IMAGES)
    print(os.getcwd()) 
except IOError as error: 
    print("Error :", error)
    
```

```python
try:
    print(x1)
except Exception as error_inst:
    print("Ha ocurrido una excepción...", error_inst)
```

```python
try:
    print(x1)
except NameError:
    print("La variable x no está definida")
except:
    print("Ha ocurrido otro tipo de error")
else:
    print("Nada ha salido bien")
```

```python
try:
    print(x1)
except:
    print("Nada ha salido bien")
finally:  # El bloque finally nos permite ejecutar código,independientemente del resultado de try/except 
    print("El 'try except' ha finalizado")
```

```python
try:
    f = open("demofile.txt")
    f.write("Lorum Ipsum")
except:
    print("El programa no  pudo guardar el archivo")
finally:  # En este caso nos permite cerrar el fichero o archivo
    f.close()
```

### os.error
os.error: todas las funciones de este módulo generan OSError en el caso de nombres de archivo y rutas de accesos no válidos o inaccesibles, u otros argumentos que tienen el tipo correcto, pero que el sistema operativo no acepta. os.error es un alias para la excepción OSError integrada.


```python
print(os.getcwd()) 
try:
    PATH_IMAGES = PATH_FILE + "\images"
    os.chdir(PATH_IMAGES)
    print(os.getcwd()) 

except IOError: 
    print("Error ")
```
