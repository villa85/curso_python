3. Escriba la función calcular_hipotenusa.  La función debe tomar la longitud de dos lados de un ángulo recto y retornar
   la longitud de la hipotenusa.

```python
# 3. escriba su respuesta aqui...
import math

def calcular_hipotenusa (cateto1, cateto2):
    hipotenusa = math.sqrt((cateto1**2) + (cateto2**2))
    print(f"La longitud de la hipotenusa es: {hipotenusa}")

first_cat = float(input("Introducir el Primer cateto: "))
second_cat = float(input("Introducir el Segundo cateto: "))

calcular_hipotenusa(first_cat, second_cat)
```

El parámetro final, switch, también debe ser opcional con un valor predeterminado de Falso. 

    La función debe multiplicar los dos enteros (n y mult_int), y si switch es True, debe cambiar el signo 
    al resultado de la multiplicación.

    Realice pruebas y muestre los resultados para diferentes valores de los parámetros opcionales.

```python
# 8. escriba su respuesta aqui...
def nums(n, mult_int = 5, swich = False):
    if if_integer(n):
        result = int(n) * mult_int
        if not swich:
            return result
        elif swich:
            result_signo = cambia_signo(result)
            return result_signo
    else:
        print(f'El valor "{n}" introducido no es un entero, por favor introduzca un número entero')


def cambia_signo(number):
    if number < 0:
        positivo = abs(number)
        return positivo
    elif number > 0:
        negativo = number * -1
        return negativo
    else:
        return number

def if_integer(string): # compruebo que el valor introducido es un entero

    if string[0] == ('-') or string[0] == ('+'):
        return string[1:].isdigit()

    else:
        return string.isdigit()

numero = input("Introduz el número a multiplicar: ")
resultado = nums(numero, swich=True)

print(resultado)
```

14. Utilizando funciones recursivas, elabora un programa que cálcule la serie de n números fibonaccis

```python
# 14. escriba su respuesta aqui...
def fibonaccis(n):
    a = b = 1
    for i in range(n):
        yield a
        a, b = b, a + b

for i in fibonaccis(10):
    print(i)
```
