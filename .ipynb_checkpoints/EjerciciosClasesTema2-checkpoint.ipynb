### Trabajando con fecha y tiempo

#### Objeto timedelta

Es la base para operaciones con fechas.  Representa una duración de tiempo, es decir, la diferencia entre dos fechas o instantes de tiempo.

    datetime.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0) este retorna;
    datetime.timedelta(days, seconds, microseconds)

Almacena internamente sólo los **días, segundos y microsegundos**, por ende, los argumentos se convierten en esas unidades:

    o	Un milisegundo se convierte en 1000 microsegundos.
    o	Un minuto se convierte en 60 segundos.
    o	Una hora se convierte en 3600 segundos.
    o	Una semana se convierte en 7 días.


```python
from datetime import timedelta

delta = timedelta(
     days=50,
     seconds=27,
     microseconds=10,  
     milliseconds=29000,   # convierte a microsegundos (29000*1000)
     minutes=5,            # convierte los min a segundos (5*60)
     hours=26,              # convierte horas a segundos (8*3600)
     weeks=2               # convierte semanas a días (2*7)
)

delta  # mostrar 
```

```python
delta = timedelta(weeks=265)
delta
```

#### Objeto datetime

El objeto datetime es un único objeto que contiene toda la información de un objeto date y un objeto time.  
Como un objeto date, datetime asume el calendario gregoriano actual extendido en ambas direcciones; como un objeto time, datetime supone que hay exactamente 3600*24 segundos en cada día.

    datetime.datetime(year, month, day, hour=0, minute=0, second=0, microsecond=0, tzinfo=None, *, fold=0)


```python
# datetime contiene toda la información de un objeto date y un objeto time

from datetime import datetime

# Métodos

mi_fecha = datetime.today()
print("Fecha actual (today): ", mi_fecha)

fecha = datetime.now()  # puede incluir la zona horaria
print("Fecha actual (now)  : ", fecha)


fechautc = datetime.utcnow() # UTC (Current time in Coordinated Universal Time (UTC) )
print("Fecha actual (utc)  : ", fechautc)
```

La biblioteca estándar de Python no contiene información de la zona horaria, necesitamos un módulo de terceros para esto; la elección habitual es pytz

```python
from datetime import datetime
from pytz import timezone

zonas = ['UTC','US/Pacific','Europe/Berlin','Australia/Adelaide','America/New_York', 'Africa/Luanda']
for zona in zonas:
    fecha = datetime.now(timezone(zona))
    print(fecha.strftime("%Y-%m-%d %H:%M:%S %Z%z"),zona)
```

```python
from zoneinfo import ZoneInfo
from datetime import datetime

# Creamos una fecha para la ciudad de México.
dt = datetime(2020, 10, 31, 13, tzinfo=ZoneInfo("America/Mexico_City"))
print(f"Fecha en Mexico {dt}")
print(f"Zona Horaria Mexico: {dt.tzname()}")

berlin = ZoneInfo('Europe/Berlin')
madrid = ZoneInfo('Europe/Paris')
losangeles = ZoneInfo('America/Los_Angeles')

# Modificamos la zona horaria
print(f"Fecha en Berlin {dt.astimezone(berlin)}")
print(f"Fecha en Madrid {dt.astimezone(madrid)}")
print(f"Fecha en Los Angeles {dt.astimezone(losangeles)}")
```

### Fechas de Formato ISO

El formato internacional definido por ISO (IS0 8601) define un sistema numérico de fechas como se muestra a continuación: AAAA-MM-DD o la representación completa: YYYY-MM-DDTHH:MM:SS (2007-11-03T24:12:03)

    Donde:
            o	AAAA es el año [todos los dígitos, p.ej. 2012]
            o	MM es el mes [01 (Enero) hasta 12 (Diciembre)]
            o	DD es el día [01 hasta 31]
            o	HH horas [00 hasta 23]
            o	MM minutos [00 hasta 59]
            o	SS segundos [00 hasta 59]


```python
from datetime import date, datetime, timezone
```

```python
from datetime import date
from datetime import datetime
from datetime import timezone
```

```python
date(2002, 12, 4).isoformat()
```

```python
datetime(2019, 5, 18, 15, 17, 8, 132263).isoformat()
```

```python
datetime.now()
```

```python
# Local to ISO 8601:
print(datetime.now().isoformat())

# UTC to ISO 8601:
print(datetime.utcnow().isoformat())
print(datetime.now().replace(microsecond=0).isoformat())  # 2021-12-14T20:50:45

# UTC to ISO 8601 with TimeZone information (Python 3):
print(datetime(2019, 5, 18, 15, 17, tzinfo=timezone.utc).isoformat())
print(datetime.now().astimezone().isoformat())
```

### Formato de Fecha timestamp

Es bastante común almacenar la fecha y la hora como una marca de tiempo en una base de datos. Una marca de tiempo Unix es el número de segundos entre una fecha determinada y el 1 de enero de 1970.


```python
from datetime import datetime
```

```python
# Convertir un timestamp a datetime

timestamp = 1639511735.84148
fecha = datetime.fromtimestamp(timestamp)

print("Timestamp:   ", timestamp)
print("Fecha:       ", fecha)
print("type(fecha): ", type(fecha))
```

```python
# Convertir un datetime a timestamp
fecha = datetime.now()

timestamp = datetime.timestamp(fecha)

print("Fecha", fecha)
print("Timestamp =", timestamp)
print("Timestamp =", type(timestamp))
```

#### TextCalendar(firstweekday=0): 

Esta clase puede ser usada para generar calendarios de texto simple.

###  Servicios de internacionalización (Módulo locale)

El módulo locale abre el acceso a la base de datos y la funcionalidad de POSIX locale (acrónimo de Portable Operating System Interface, y X viene de UNIX como señal de identidad de la API) . El mecanismo POSIX locale permite a los programadores tratar ciertos problemas culturales en una aplicación, sin requerir que el programador conozca todos los detalles de cada país donde se ejecuta el software.

    locale.setlocale(category, locale=None)

Si locale está dado y no es None, setlocale() modifica la configuración de localización para category. Las categorías disponibles son:

    o	locale.LC_ALL todas las categorías. 
    o	locale.LC_NUMERIC configuraciones de números. 
    o	locale.LC_TIME configuraciones de fechas. 
    o	locale.LC_MONETARY configuraciones de moneda. 
    o	locale.LC_COLLATE configuraciones de comparación de textos. 
    o	locale.LC_CTYPE configuraciones de caracteres.

```python
x = random.getstate()
print(x)
```

random.setstate(state): El state debería haberse obtenido de una llamada previa a getstate(), y setstate() reestablece el estado interno del generador al que tenía cuando se llamó a la función getstate().

```python
print(random.random()) # imprimir un número random
print(random.random())

state = random.getstate() # captura el state:

print(random.random()) # imprime otro número random

random.setstate(state) # restablece el state:

#El siguiente número random debería ser igual al anterior
print(random.random())

# cada vez que genera un número aleatorio, actualiza el generador,
# al reestablecerlo vuelve a generar el mismo.  
# Si generamos otro es diferente

print(random.random())
```

numpy.random.rand
numpy.random.rand(d0, d1, ..., dn)
Random values in a given shape.

Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).

```python
import numpy as np      # crear arreglos
import matplotlib.pyplot as plt


def twospirals(n_points, noise=.9):
    """
     Returns the two spirals dataset.
    """
    n = np.sqrt(np.random.rand(n_points,1)) * 540 * (2*np.pi)/360
    d1x = (-np.cos(n)*n + np.random.rand(n_points,1) * noise)/10
    d1y = (np.sin(n)*n + np.random.rand(n_points,1) * noise)/10
    return (np.vstack((np.hstack((d1x,d1y)),np.hstack((-d1x,-d1y)))), 
            np.hstack((np.zeros(n_points),np.ones(n_points))))


X, y = twospirals(100)
```

```python
plt.rcParams['figure.figsize'] = (6,6) 
plt.style.use('seaborn-white') 

fig = plt.figure()
ax1 = fig.add_subplot()
plt.title('Training Set',fontsize = 20 )

ax1.scatter(X[y==0,0], X[y==0,1], label='class 1', marker='o', cmap='viridis',alpha = 0.5, s = 100)
ax1.scatter(X[y==1,0], X[y==1,1], label='class 2', marker='o', cmap='viridis',alpha = 0.5, s = 100)

ax1.set_xlabel('x',fontsize=18,fontstyle='italic')
ax1.set_ylabel('y',fontsize=18,fontstyle='italic')
```

### Cuando utilizar search() y match()

    1.	Si desea ubicar una coincidencia en cualquier lugar de la cadena, use search().
    2.	Si necesita hacer coincidir al principio de la cadena, o para hacer coincidir toda la cadena, use match. Es más rápido.


```python
string_with_newlines = """something
someotherthing"""

import re

print("1. ", re.match('some',       string_with_newlines)) # matches
print("2. ", re.match('someother',  string_with_newlines) )# won't match
print("3. ", re.match('^someother', string_with_newlines, re.MULTILINE) )# also won't match
print("4. ", re.search('someother', string_with_newlines)) # finds something
print("5. ", re.search('^someother', string_with_newlines, re.MULTILINE) )# also finds something
print("---------")

m = re.compile('thing$', re.MULTILINE)

print("6. ",m.match(string_with_newlines)) # no match
print("7. ",m.match(string_with_newlines, pos=4)) # matches
print("8. ",m.search(string_with_newlines, re.MULTILINE)) # also matches
```

#### Inicio y fin de la extracción: ()
Los paréntesis ( ) no forman parte del patrón a comprobar, pero indican respectivamente dónde empieza y termina la extracción del texto. 

```python
frase = "Tengo dos correos electrónicos que son m.sierra@cloudftic.com y maria.c.sierra.f@gmail.com"
patron = '@([^ ]*)' # comienza por @, no incluye blancos, 0 o mas caracteres
re.findall(patron, frase)
```

```python
patron = '([^ ]*)@' # termina en @, no incluye blancos, 0 o mas caracteres
re.findall(patron, frase)
```

```python
patron = re.compile("^([a-z0-9_.-]+)@([\da-z.-]+).([a-z.]{2,6})$")

if patron.search("maria.c.sierra.f@gmail.com"): # Comprobemos sí este es un correo electronico valido
    print("Correo valido!")
else:
    print("Correo Invalido!")
```

```python
# Para establecer secuencias alternativas podemos usar la barra vertical '│'. 
# En el siguiente ejemplo comprobamos diversas extensiones de ficheros:
import re
extensiones = ['jpg ', 'png', 'gif', 'mp3', 'doc']

for tipoarchivo in extensiones:
    
    if re.match('jpg|png|gif|bmp', tipoarchivo):
        print('La extensión ', tipoarchivo, ' se corresponde con una imagen.')
    else:
        print('La extensión ', tipoarchivo, ' no se corresponde con una imagen.')

```

Mostrar "Su archivo o fichero es un archivo de texto" si la extensión es 
(.txt, .doc, .docx, .odt, .sxw)

```python
#Si tenemos que casar secuencias alternativas por el medio de un texto debemos delimitar con paréntesis las alternativas:
import re
fichero = input('Introduzca un nombre de fichero: ')
if re.match(r'.+\.(jpg|png|gif|bmp)', fichero):
    print('El fichero es una imagen.')
else:
    print("No es una imagen")
```
