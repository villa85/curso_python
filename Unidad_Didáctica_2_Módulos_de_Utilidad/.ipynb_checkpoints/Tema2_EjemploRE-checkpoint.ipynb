## Ejemplos de Expresiones Regulares

```python
import os
import re
```

```python
# Extraer los números

linea = "well-formed 1459 sentences 8262 in 1973 this 150 new language 1a"
re.findall('[0-9]+',linea ) # * . + ?
```

```python
# Extraer los números

linea = "well-formed 1459 sentences 8262 in 1973 this 15 new language."
re.findall('[0-9]+',linea )
```

```python
# Extraer Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL
linea = "as JavaScript or C++.  The new programming  language  Python has very different"
re.findall('[A-Z][a-z]+',linea )
```

```python
linea = "as JavaScript, Perl, Ruby or C++.  The new programming  language  Python has very different"
re.findall('[J][a-z]+',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('[J|R|B|P|F][a-z]+',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('\ [J|R|B|P|F][a-z]+',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('C\++',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('C\+.',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('[P][a-z]+[A-Z][a-z]+',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('[A-Z][a-z]+[A-Z][a-z]+',linea )
```

```python
linea = "as Java, C++, PHP, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('[P][A-Z]+',linea )
```

```python
linea = "as Java, C++, PHP, CPU, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('[P|C][A-Z]+',linea )
```

```python
linea = "as Java, C++, PHP, CPU, ONU, Onu, Ruby, Basic, Perl, JavaScript, Python, PowerBuilder, Fortran, COBOL.  The new programming  language  Python has very different"
re.findall('[A-Z][A-Z]+',linea )
```

```python
linea = "This file was created by the University of Michigan for the Python3 course https://umich.edu/, http://www.google.com"
re.findall("http[s]?://(?:[a-zA-Z]|[0-9]|[$_@.&+]|[!*\(\),]|(?:%[0-9a-z-A-z][0-9a-zA-F]))+",linea)      
```

```python
linea = "This file was created by the University of Michigan for the Python3 course https://umich.edu/, http://www.google.com"
re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+',linea)
```

```python
# Cuales expresiones voy a utilizar?

re.findall('\ [J|R|B|P|F][a-z]+',linea ) # re.findall('\ [J|R|B|P|F][a-z]+',linea )
re.findall('C\++',linea ) # ['C++']
re.findall('[A-Z][a-z]+[A-Z][a-z]+',linea )
re.findall('[A-Z][A-Z]+',linea )  # ['PHP', 'CPU', 'COBOL']
re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+',linea) # ['https://umich.edu']

```

```python

#PATH_FILES = r"C:\Users\msierra\Desktop\My Curso Python\Proyectos\data" # absoluta
#os.chdir(PATH_FILES)

my_file = open("data/regex_ejemplo.txt") #relativa
contar = 0

for linea in my_file:  # Estamos leyendo linea por linea
    linea = linea.rstrip()   # limpiamos los espacios en blanco al final
    encontre = re.findall('[J|R|B|P|F][a-z]+',linea) 
    if len(encontre) != 0:
        print(encontre)
        contar = contar + len(encontre)
    encontre = re.findall('C\++',linea ) # ['C++']
    if len(encontre) != 0:
        contar = contar + len(encontre)
    encontre = re.findall('[P][a-z]+[A-Z][a-z]+',linea ) #['PowerBuilder']
    if len(encontre) != 0:
        contar = contar + len(encontre)            
    encontre = re.findall('[J][a-z]+[A-Z][a-z]+',linea ) # ['JavaScript']
    if len(encontre) != 0:
        contar = contar + len(encontre)            
print(contar)   
```
